### Analysis

- Over the course of (45) days, (249) mice were subjected to (4) different treatment regimens. The purpose of the study was to compare the effectiveness of Pymaceuticals' drug of interest, Capomulin, against other treatment regimens. 

- At the conclusion of the study, it was found that Capomulin is an effective treatment regimen for shrinking tumors in mice.

- Out of the (4) treatment regimens in question (Capomulin, Ramicane, Infubinol, and Ceftamin), Capomulin was found to be one best drugs for shrinking tumors in mice, with very similar results to the drug Ramicane. 

- There appears to be a Positive Correlation (+0.84) between Mouse Weight (g) and Average Tumor Volume when looking at data for the Capomulin regime. The Regression line indicates that weight can influence the Average Tumor Volume in mice.

- With one of the lowest values in Standard Deviation & Standard Error in Tumor Volume, the team has a higher confidence in the results of Capomulin than the majority of other treatments. The treatment regimen, Ramicane, is the only drug with lower values in these categories. 

- Although results for Capomulin show promise, the treatment regimen Ramicane appears to be the most effective drug out of the (4) treatments in question. 
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "./data/Mouse_metadata.csv"
study_results_path = "./data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data_df = pd.merge(study_results,mouse_metadata,how="left",on="Mouse ID")

# Display the data table for preview
combined_data_df.head()
```

```python
# Checking the number of mice.
len(combined_data_df["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mice_id_duplicates = combined_data_df[combined_data_df.duplicated(subset=["Mouse ID", "Timepoint"])]["Mouse ID"].unique()
mice_id_duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_id_data = combined_data_df[combined_data_df["Mouse ID"] == "g989"]
duplicate_mouse_id_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_combined_data_df = combined_data_df[combined_data_df["Mouse ID"].isin(mice_id_duplicates) == False]
cleaned_combined_data_df
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mice_count = len(cleaned_combined_data_df["Mouse ID"].unique())
cleaned_mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean = cleaned_combined_data_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median = cleaned_combined_data_df.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
variance = cleaned_combined_data_df.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
std_dev = cleaned_combined_data_df.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
sem = cleaned_combined_data_df.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

summary_statistics = pd.DataFrame({
    "Mean Tumor Volume": mean,
    "Median Tumor Volune": median,
    "Tumor Volume Variance": variance,
    "Tumor Volume Std. Dev.": std_dev,
    "Tumor Volume Std. Err.": sem

})

summary_statistics
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
aggregated_summary_statistics = cleaned_combined_data_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std", "sem"]})
aggregated_summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint = cleaned_combined_data_df.groupby(["Mouse ID"])["Timepoint"].max()
greatest_timepoint = greatest_timepoint.reset_index()
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_df = greatest_timepoint.merge(cleaned_combined_data_df, on=["Mouse ID", "Timepoint"], how="left")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_list=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatment_list:
    
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volumes = merged_df.loc[merged_df["Drug Regimen"]==treatment,"Tumor Volume (mm3)"]
    
    # add subset 
    tumor_volume_list.append(tumor_volumes)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volumes.quantile([0.25,0.5,0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    lowerbound = lowerq - (1.5 * iqr)
    upperbound = upperq + (1.5 * iqr)
    
    outliers = tumor_volumes.loc[(tumor_volumes < lowerbound) | (tumor_volumes > upperbound)]
    print(f"The potential outliers for {treatment} are {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
tumor_boxplot = dict(markerfacecolor='red', markersize = 15)
plt.boxplot(tumor_volume_list, labels = treatment_list, flierprops=tumor_boxplot)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_df = cleaned_combined_data_df[cleaned_combined_data_df["Drug Regimen"] == "Capomulin"]
capomulin_mice = capomulin_df[capomulin_df["Mouse ID"] == "l509"]
plt.plot(capomulin_mice["Timepoint"],capomulin_mice["Tumor Volume (mm3)"])
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

capomulin_df = cleaned_combined_data_df[cleaned_combined_data_df["Drug Regimen"] == "Capomulin"]
capomulin_avg = capomulin_df.groupby(["Mouse ID"]).mean()

avg_tumor_volume = capomulin_avg["Tumor Volume (mm3)"]
avg_mouse_weight = capomulin_avg["Weight (g)"]


plt.scatter(avg_mouse_weight,avg_tumor_volume,marker="o",facecolors="blue")
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
capomulin_correlation = st.pearsonr(avg_mouse_weight,avg_tumor_volume)
print(f"The correlation between mouse weight and the average tumor volume is {round(capomulin_correlation[0],2)}")

regression_model = st.linregress(avg_mouse_weight,avg_tumor_volume)
slope = regression_model[0]
intercept = regression_model[1]
y_values = avg_mouse_weight * slope + intercept


plt.scatter(avg_mouse_weight,avg_tumor_volume)
plt.plot(avg_mouse_weight,y_values,color="red")
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```
